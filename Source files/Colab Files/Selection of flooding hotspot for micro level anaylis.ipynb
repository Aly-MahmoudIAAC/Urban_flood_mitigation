#Total Bounding box area for neighbourhood area

```python
# Import necessary libraries
import math

# Constants
R = 6371000  # Earth's radius in meters


lat1, lon1 = 19.0438, 72.8398
lat2, lon2 = 19.1148, 72.8828

# Calculate latitude and longitude differences
lat_diff = lat2 - lat1
lon_diff = lon2 - lon1

# Mean latitude
mean_lat = (lat1 + lat2) / 2

# Convert latitude difference to meters (North-South distance)
north_south_distance = (lat_diff * math.pi / 180) * R

# Convert longitude difference to meters (East-West distance)
east_west_distance = (lon_diff * math.pi / 180) * R * math.cos(mean_lat * math.pi / 180)

# Calculate the area of the bounding box in square meters
bounding_box_area = north_south_distance * east_west_distance

# Print the results
print(f"North-South Distance: {north_south_distance:.2f} meters")
print(f"East-West Distance: {east_west_distance:.2f} meters")
print(f"Bounding Box Area: {bounding_box_area:.2f} square meters")


```
